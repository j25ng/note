```python
import pandas as pd
```

```python
d = pd.read_parquet('~/code/de32-kca/data_kca/')

cols = [
        'movieCd', #영화의 대표코드를 출력합니다.
        'movieNm', #영화명(국문)을 출력합니다.
        'load_dt', # 입수일자
        'repNationCd', #한국외국영화 유무
    ]

df = d[cols]

df_G = df[(df['load_dt'] == 20180103) & (d['repNationCd'] == 'G')]
df_G
```

```python
df_K = df[(df['load_dt'] == 20180103) & (d['repNationCd'] == 'K')]
df_K
```

```python
df_m = df_G.merge(df_K, on=['movieCd', 'movieNm', 'load_dt'], how='left')
df_m
```

```python
df_d = df_m.drop(columns='repNationCd_x').reset_index(drop=True)
```

```python
df_d
```

```python
df_N = df_d[df_d['repNationCd_y'].notnull()]
df_N = df_N.rename(columns={'repNationCd_y': 'nationCd'})
df_N
```
